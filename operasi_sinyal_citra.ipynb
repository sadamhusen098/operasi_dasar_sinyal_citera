Notebook lengkap Bagian A–D sesuai instruksi tugas.

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import cv2
```

## Bagian A - Operasi pada Sinyal 1D

```python
# A1 Membuat sinyal
n = np.arange(0,31)
x1 = np.sin(0.5*np.pi*n)
x2 = np.where(n < 5, 0, 1)
plt.figure(figsize=(10,4))
plt.stem(n,x1)
plt.title('x1[n]')
plt.show()
plt.figure(figsize=(10,4))
plt.stem(n,x2)
plt.title('x2[n]')
plt.show()
```

```python
# A2 Penjumlahan
y = x1 + x2
plt.figure(figsize=(10,4))
plt.stem(n,y)
plt.title('y[n] = x1[n] + x2[n]')
plt.show()
```

```python
# A3 Penggeseran
for k in [-3,0,4]:
    plt.figure(figsize=(8,3))
    plt.stem(n, np.sin(0.5*np.pi*(n-k)))
    plt.title(f'Shift k={k}')
    plt.show()
```

```python
# A4 Amplifikasi
for a in [0.5,1,2,-1]:
    plt.figure(figsize=(8,3))
    plt.stem(n,a*x1)
    plt.title(f'Alpha={a}')
    plt.show()
```

## Bagian B - Operasi pada Citra 2D

```python
# Ganti dengan file citra Anda
img = cv2.imread('images/citra_yang_digunakan.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
print('Ukuran:', img.shape)
print('Tipe:', img.dtype)
print('Min:', img.min())
print('Max:', img.max())
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
# B2 Penjumlahan citra
img2 = cv2.flip(img,1)
add = cv2.add(img,img2)
plt.imshow(add)
plt.axis('off')
plt.show()
```

```python
# B3 Penggeseran citra
rows, cols = img.shape[:2]
for dx,dy in [(30,0),(0,30),(30,30)]:
    M = np.float32([[1,0,dx],[0,1,dy]])
    shifted = cv2.warpAffine(img,M,(cols,rows))
    plt.imshow(shifted)
    plt.axis('off')
    plt.show()
```

```python
# B4 Amplifikasi citra
for a in [0.5,1,1.5,2]:
    amp = np.clip(img*a,0,255).astype(np.uint8)
    plt.imshow(amp)
    plt.title(f'Alpha={a}')
    plt.axis('off')
    plt.show()
```

## Bagian C - Uji Sistem Linier

```python
# C1 Homogenitas
x = x1
for a in [0.5,2,3]:
    left = 2*(a*x)
    right = a*(2*x)
    print(a, np.allclose(left,right))
```

```python
# C2 Additivitas
left = 2*(x1+x2)
right = 2*x1 + 2*x2
print(np.allclose(left,right))
```

```python
# C3 Sistem linier vs non-linier
T1 = 2*x1
T2 = x1**2
plt.plot(n,T1,label='T1=2x')
plt.plot(n,T2,label='T2=x^2')
plt.legend()
plt.show()
```
